```python
import APALD
from sklearn.datasets import make_blobs
from tqdm import tqdm
import timeit
import matplotlib.pyplot as plt
```

```python
sizes = [1000, 2500, 10000, 25000, 100000]
min_times = []
for n in tqdm(sizes):
    data, labels = make_blobs(n_samples=n, n_features=1024, centers=100)
    times = timeit.repeat(lambda: APALD.APALD().fit_predict(data), number=1, repeat=5)
    min_times.append(min(times))
```

```python
plt.plot(sizes, min_times)
```
